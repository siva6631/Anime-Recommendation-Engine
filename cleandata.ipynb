```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('cleananimedata.csv')
df
```

```python
df=df.drop(columns='Unnamed: 0',axis=1)
df
```

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
ps=PorterStemmer()
wl=WordNetLemmatizer()
```

```python
df['title']=df['name'].astype(str)+" "+df['genre'].astype(str)+" "+df['type'].astype(str)
```

```python
df['title'].head().tolist()
```

```python
type('title')
```

```python
corpus=[]
for i in range(len(df)):
    rp=re.sub('[^a-zA-Z]',' ',df['title'][i])
    rp=rp.lower()
    rp=rp.split()
    rp=[ps.stem(word)for word in rp if not word in set(stopwords.words('english'))]
    rp=' '.join(rp)
    corpus.append(rp)
corpus
```

```python
cv=CountVectorizer()
tf=TfidfVectorizer()
```

```python
#vector encoding
```

```python
vector=cv.fit_transform(corpus).toarray()
vector
```

```python
vector.shape
```

```python
#cosinesimilarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
#observing highest similarity score
```

```python
sorted(similarity[0],reverse=True)
```

```python
#creating a function 
```

```python
def recommend_engine(movie):
    movie_title=df[df['name']==movie].index[0]
    sim=similarity[movie_title]
    movie_name=sorted(enumerate(sim),reverse=True,key=lambda x:x[1])[1:7]
    for i in movie_name:
        print(df.iloc[i[0]]['name'])
```

```python
recommend_engine('Steins;Gate')
```

```python
recommend_engine('Mushishi')
```
